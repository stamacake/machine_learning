```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tr = pd.read_csv('../input/mlcourse-2019-mai-autumn/train.csv').copy()
# Матрица Объекты-признаки
X = tr[['GrLivArea','YearRemodAdd','LotArea']]
# Матрица Ответов
y = tr['SalePrice']

X_scaled_1 = (X['GrLivArea'] - X['GrLivArea'].mean())/X['GrLivArea'].std()
X_scaled_2 = (X['YearRemodAdd'] - X['YearRemodAdd'].mean())/X['YearRemodAdd'].std()
X_scaled_3 = (X['LotArea'] - X['LotArea'].mean())/X['LotArea'].std()
# Задаем начальные приближения для w0 и w1
w0 = float(0.01)
w1 = float(0.01)
w2 =float(0.03)
w3 = 0.02
L = 0.000075 #шаг обновления весов
epochs = 200 # кол-во итераций
print(L)
mse = 0.4
lmbd = 2


print('lmbd = ',lmbd)
n = float(len(X_scaled_1))
    
for i in range(epochs): 
    w0 = w0+L*sum((y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3+lmbd*np.sign(w0)))
    w1 = w1+L*sum(X_scaled_1*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_2*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_3*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+lmbd*np.sign(w1)) 
    w2 = w2+L*sum(X_scaled_1*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_2*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_3*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+lmbd*np.sign(w2))
    w3 = w3+L*sum(X_scaled_1*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_2*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+X_scaled_3*(y - w0 - w1*X_scaled_1-w2*X_scaled_2-w3*X_scaled_3)+lmbd*np.sign(w3))
   
    
    plt.show()
    
print (w0, w1, w2, w3)

mse = 0
ee=0;
for i in range(y.shape[0]):
    
    if(y[i]>(w0+w1*X_scaled_1[i]+w2*X_scaled_2[i]+w3*X_scaled_3[i])):
        ee = (w0+w1*X_scaled_1[i]+w2*X_scaled_2[i]+w3*X_scaled_3[i])/(y[i])
    else:
        ee = (y[i])/(w0+w1*X_scaled_1[i]+w2*X_scaled_2[i]+w3*X_scaled_3[i])
    mse += ee
    #print('mse =', mse)
print('mse = ',1-mse/y.shape[0])
    


print('g')
ts1 = pd.read_csv('../input/mlcourse-2019-mai-autumn/test.csv').copy()
# Матрица Объекты-признаки
X1 = ts1['GrLivArea']
X2 = ts1['YearRemodAdd']
X3 = ts1['LotArea']



Y1 = (X1 - X1.mean())/X1.std()
Y2 = (X2 - X2.mean())/X2.std()
Y3 = (X3 - X3.mean())/X3.std()

#for i in range(ts1.shape[0]):
 #   print(i+782, w0+w1*Y1[i]+w2*Y2[i]+w3*Y3[i])
    
import csv
with open('submission2_v5.csv', 'w', newline='') as f:
    fieldnames = ['Id', 'SalePrice']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(ts1.shape[0]):
        writer.writerow({'Id': i+782,'SalePrice': w0+w1*Y1[i]+w2*Y2[i]+w3*Y3[i]})    




print(1-mse)
```
